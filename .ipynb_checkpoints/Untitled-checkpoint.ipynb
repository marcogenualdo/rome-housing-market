```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.neighbors import KernelDensity
from sklearn.model_selection import GridSearchCV

# generating data
data = np.concatenate([np.random.randn(200) * 2, np.random.randn(200) + 10.])

# kde density estimation with cross validated bandwidth
kde_bands = np.linspace(0.1, 2, 20)
params = {'bandwidth': kde_bands}
grid = GridSearchCV(KernelDensity(), params, cv=10)
grid.fit(data.reshape(-1,1))

# plotting cv scores and density
fig, ax = plt.subplots(1,2, figsize=(15,5))
ax[0].plot(kde_bands, grid.cv_results_['mean_test_score'])
ax[0].title.set_text('Cross Validation Score')

x = np.linspace(-5,20,100)
y = np.exp(grid.best_estimator_.score_samples(x.reshape(-1,1)))
ax[1].plot(x,y)
ax[1].title.set_text('Predicted density')

plt.show()
print('Best bandwidth:', grid.best_estimator_.bandwidth)
```
